## 1.4 Data Frame

```python
src = "./data/src/"
dest = "./data/cleaned/"

fn = "0-antes.csv"
```

```python
df = pd.read_csv(src+fn)
df
```

```python
df["BB"] =  df["BB"].str.replace("BB", "").astype(int)
df.head()
```

```python
df.index = df["BB"]
df.drop(columns="BB", inplace=True)
df.head()
```

```python
df.columns = [i.lower().strip() for i in df.columns]
df
```

```python
for col in df.columns : 
    df[col] = df[col].str.replace("\n", ", ")

df
```

```python
LI = []
for col in df.columns : 
    LI.append( df[col].apply(lambda i : i.split(",")))

LI = pd.DataFrame(LI)
LI.T
```

```python
for col in df.columns : 
    df[col]=  df[col].apply(lambda i : i.split(","))
df
```

```python
LI = []
for col in df.columns : 
    LI.append( df[col].apply(lambda i : len(i)))

LI = pd.DataFrame(LI)
LI.T

```

```python
def select(li) : 
    if len(li)==2 : 
        return li[1].strip()
    return li

for col in df.columns : 
    df[col] = df[col].apply(lambda i : select(i))
df
```

```python
df.iloc[0,0] = "2x+"
```

```python
for col in df.columns : 
    df[col] = df[col].apply(lambda i : i.replace(" ", ",").replace(" ", ""))
df
```

```python
for col in df.columns : 
    df[col] = df[col].str.lower()
df
```

```python
df.to_csv(dest+fn)
```
